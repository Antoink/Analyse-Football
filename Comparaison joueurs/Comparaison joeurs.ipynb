Comparaison de Joueurs sur la saison 2021-2022

```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Radar, FontManager

df = pd.read_excel("/Users/Utilisateur/Documents/Comparaison joueurs/2021-2022 Football Player Stats.xlsx")
df_min= df[df["Min"]>=900]

défenseurs = ["DF", "MFDF", "DFMF" ]
df_defenseurs= df_min[df_min["Pos"].isin(défenseurs)]

milieux = ["MF", "DFMF", "MFFW", "MFDF", "FWMF"]
df_milieux = df_min[df_min["Pos"].isin(milieux)]

attaquants= ["FW", "MFFW", "FWMF"]
df_attaquants= df_min[df_min["Pos"].isin(attaquants)]
```

Défenseurs

```python
colonnes_defenseurs_lateraux = ["Tkl", "TklWon", "TklDri", "Crs", "PasTotCmp%", "PasTotAtt", "Assists", "AerWon%", "DriSucc%"]
dfdeflaterauxstats = df_defenseurs[colonnes_defenseurs_lateraux]


moyennes = dfdeflaterauxstats[colonnes_defenseurs_lateraux].mean()
minimums = dfdeflaterauxstats[colonnes_defenseurs_lateraux].min()
maximums = dfdeflaterauxstats[colonnes_defenseurs_lateraux].max()


def create_radar_chart(columns, label):
    moyennes = dfdeflaterauxstats[columns].mean()
    minimums = dfdeflaterauxstats[columns].min()
    maximums = dfdeflaterauxstats[columns].max()
    maxgraph = maximums 
    
    radar = Radar(
        params=columns,
        min_range=minimums.values, 
        max_range=maxgraph.values,
        round_int=[False] * len(columns),
        num_rings=4,
        ring_width=1,
        center_circle_radius=1
    )
    
    fig, ax = radar.setup_axis()
    rings_inner = radar.draw_circles(ax=ax, facecolor='#ffffff', edgecolor='#add8e6')
    radar_output = radar.draw_radar(
        moyennes.values, 
        ax=ax,
        kwargs_radar={'facecolor': '#4682b4', 'alpha': 0.6},
        kwargs_rings={'facecolor': '#add8e6', 'alpha': 0.4}
    )
    radar_poly, rings_outer, vertices = radar_output

    range_labels = radar.draw_range_labels(
        ax=ax, 
        fontsize=15, 
        zorder=2.5, 
        fontproperties=roboto_thin.prop, 
        color='#080561'
    )

    param_labels = radar.draw_param_labels(
        ax=ax, 
        fontsize=15, 
        fontproperties=roboto_thin.prop, 
        color='#040c0f'
    )

    lines = radar.spoke(
        ax=ax, 
        color='#040c0f', 
        linestyle=':', 
        zorder=2,
        alpha=0.1
    )

    plt.title(label)
    plt.show()

URL1 = ('https://raw.githubusercontent.com/googlefonts/SourceSerifProGFVersion/main/fonts/'
        'SourceSerifPro-Regular.ttf')
serif_regular = FontManager(URL1)
URL2 = ('https://raw.githubusercontent.com/googlefonts/SourceSerifProGFVersion/main/fonts/'
        'SourceSerifPro-ExtraLight.ttf')
serif_extra_light = FontManager(URL2)
URL3 = ('https://raw.githubusercontent.com/google/fonts/main/ofl/rubikmonoone/'
        'RubikMonoOne-Regular.ttf')
rubik_regular = FontManager(URL3)
URL4 = 'https://raw.githubusercontent.com/googlefonts/roboto/main/src/hinted/Roboto-Thin.ttf'
roboto_thin = FontManager(URL4)
URL5 = ('https://raw.githubusercontent.com/google/fonts/main/apache/robotoslab/'
        'RobotoSlab%5Bwght%5D.ttf')
roboto_bold = FontManager(URL5)

create_radar_chart(colonnes_defenseurs_lateraux, "Lateraux")


colonnes_defenseurs_centraux = ["Tkl", "TklWon", "Int", "Clr", "AerWon%", "Blocks", "Err", "Recov"]
dfdefcentrauxstats = df_defenseurs[colonnes_defenseurs_centraux]

moyennes = dfdefcentrauxstats[colonnes_defenseurs_centraux].mean()
minimums = dfdefcentrauxstats[colonnes_defenseurs_centraux].min()
maximums = dfdefcentrauxstats[colonnes_defenseurs_centraux].max()

def create_radar_chart(columns, label):
    moyennes = dfdefcentrauxstats[columns].mean()
    minimums = dfdefcentrauxstats[columns].min()
    maximums = dfdefcentrauxstats[columns].max()
    maxgraph = maximums 
    
    radar = Radar(
        params=columns,
        min_range=minimums.values, 
        max_range=maxgraph.values,
        round_int=[False] * len(columns),
        num_rings=4,
        ring_width=1,
        center_circle_radius=1
    )
    
    fig, ax = radar.setup_axis()
    rings_inner = radar.draw_circles(ax=ax, facecolor='#ffffff', edgecolor='#add8e6')
    radar_output = radar.draw_radar(
        moyennes.values, 
        ax=ax,
        kwargs_radar={'facecolor': '#4682b4', 'alpha': 0.6},
        kwargs_rings={'facecolor': '#add8e6', 'alpha': 0.4}
    )
    radar_poly, rings_outer, vertices = radar_output

    range_labels = radar.draw_range_labels(
        ax=ax, 
        fontsize=15, 
        zorder=2.5, 
        fontproperties=roboto_thin.prop, 
        color='#080561'
    )

    param_labels = radar.draw_param_labels(
        ax=ax, 
        fontsize=15, 
        fontproperties=roboto_thin.prop, 
        color='#040c0f'
    )

    lines = radar.spoke(
        ax=ax, 
        color='#040c0f', 
        linestyle=':', 
        zorder=2,
        alpha=0.1
    )

    plt.title(label)
    plt.show()

create_radar_chart(colonnes_defenseurs_centraux, "Centraux")
```

```python
joueur = "Ousmane Dembélé"

# Filtrer les données du joueur à comparer
dfjoueuracomparer = df_min[df_min["Player"] == joueur]

colonnes_attaquants_de_pointe = ["Goals", "Shots", "SoT%", "Assists", "G/Sh", "GcaSh", "ScaSh", "ScaFld", "SCA"]

def create_radar_chart_with_joueur(columns, label):
    moyennes = dfattaquantsstats[columns].mean()
    minimums = dfattaquantsstats[columns].min()
    maximums = dfattaquantsstats[columns].max()
    maxgraph = maximums

    # Vérifier si les données du joueur existent
    if dfjoueuracomparer.empty:
        raise ValueError(f"Les statistiques pour {joueur} ne sont pas disponibles.")
    
    # Statistiques du joueur
    joueur_stats = dfjoueuracomparer[columns].iloc[0]

    # Configuration du radar
    radar = Radar(
        params=columns,
        min_range=minimums.values,
        max_range=maxgraph.values,
        round_int=[False] * len(columns),
        num_rings=4,
        ring_width=1,
        center_circle_radius=1
    )
    
    fig, ax = radar.setup_axis()
    rings_inner = radar.draw_circles(ax=ax, facecolor='#ffffff', edgecolor='#add8e6')
    
    # Tracer les moyennes
    radar_output_avg = radar.draw_radar(
        moyennes.values,
        ax=ax,
        kwargs_radar={'facecolor': '#4682b4', 'alpha': 0.6},
        kwargs_rings={'facecolor': '#add8e6', 'alpha': 0.4}
    )
    
    # Tracer les statistiques du joueur
    radar_output_joueur = radar.draw_radar(
        joueur_stats.values,
        ax=ax,
        kwargs_radar={'facecolor': '#ff4500', 'alpha': 0.6},
        kwargs_rings={'facecolor': '#ffa07a', 'alpha': 0.4}
    )

    range_labels = radar.draw_range_labels(
        ax=ax,
        fontsize=15,
        zorder=2.5,
        fontproperties=roboto_thin.prop,
        color='#080561'
    )
    
    param_labels = radar.draw_param_labels(
        ax=ax,
        fontsize=15,
        fontproperties=roboto_thin.prop,
        color='#040c0f'
    )

    lines = radar.spoke(
        ax=ax,
        color='#040c0f',
        linestyle=':',
        zorder=2,
        alpha=0.1
    )
    
    # Ajouter une légende
    ax.legend(['Moyennes', joueur], loc='upper right', fontsize=12, prop=roboto_thin.prop)
    
    # Titre
    plt.title(label)
    plt.show()

# Exemple d'utilisation
create_radar_chart_with_joueur(
    columns=colonnes_attaquants_de_pointe,
    label=f" {joueur}"
)
```

Comparer 2 joueurs entre eux 

```python
joueur1 = "Raphaël Varane"
joueur2 = "Dayot Upamecano"

dfjoueur1 = df_min[df_min["Player"] == joueur1]
dfjoueur2 = df_min[df_min["Player"] == joueur2]



#a changer selon le poste des joueurs 
colonnecomparaison = colonnes_defenseurs_centraux
dfcomparaison = dfdefcentrauxstats

def create_radar_chart_two_players(columns, label1, label2):
    stats_joueur1 = dfjoueur1[columns].iloc[0]
    stats_joueur2 = dfjoueur2[columns].iloc[0]
    minimums = dfcomparaison[columns].min()
    maximums = dfcomparaison[columns].max()
    
    radar = Radar(
        params=columns,
        min_range=minimums.values,
        max_range=maximums.values,
        round_int=[False] * len(columns),
        num_rings=4,
        ring_width=1,
        center_circle_radius=1
    )
    
    fig, ax = radar.setup_axis()
    radar.draw_circles(ax=ax, facecolor='#ffffff', edgecolor='#add8e6')
    radar.draw_radar(stats_joueur1.values, ax=ax, kwargs_radar={'facecolor': '#ff4500', 'alpha': 0.6}, kwargs_rings={'facecolor': '#ffa07a', 'alpha': 0.4})
    radar.draw_radar(stats_joueur2.values, ax=ax, kwargs_radar={'facecolor': '#4682b4', 'alpha': 0.6}, kwargs_rings={'facecolor': '#add8e6', 'alpha': 0.4})
    radar.draw_range_labels(ax=ax, fontsize=15, zorder=2.5, fontproperties=roboto_thin.prop, color='#080561')
    radar.draw_param_labels(ax=ax, fontsize=15, fontproperties=roboto_thin.prop, color='#040c0f')
    radar.spoke(ax=ax, color='#040c0f', linestyle=':', zorder=2, alpha=0.1)
    ax.legend([label1, label2], loc='upper right', fontsize=12, prop=roboto_thin.prop)
    plt.title(f"Comparaison : {label1} vs {label2}")
    plt.show()

create_radar_chart_two_players(
    columns=colonnecomparaison,
    label1=joueur1,
    label2=joueur2
)
```
